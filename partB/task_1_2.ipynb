# Task 1.2 — Key Assumptions

**Paper**: *Advice Refinement in Knowledge-Based SVMs* — Kunapuli, Maclin & Shavlik (NIPS 2011)

Below are four assumptions that the arkSVM method specifically depends on. These are not generic ML assumptions — each one is baked into how the algorithm is formulated and would cause real problems if violated.

### Assumption 1

**Assumption:** The method assumes we are drawing a straight line (a linear boundary) to separate our data classes.

**Why the method needs it:** The whole math setup of arkSVM only works if we're dealing with flat spaces and straight lines. The expert advice is also fed into the system using straight-line boundaries (polyhedrons). If we tried to let the model draw curvy boundaries, the advice adjustment system ($F_i$ and $f_i$) would mathematically break down.

**Violation scenario:** Imagine our data classes are shaped like a donut (one class is the hole, the other is the dough). You can't draw a straight line to separate a donut. No matter how much we tweak the expert's linear advice, the method would totally fail here.

**Paper reference:** Section 2 starts by clearly stating "We wish to learn a linear classifier". Section 5 also mentions extending this to non-linear shapes as "future work".

### Assumption 2

**Assumption:** Expert advice can be mapped out using straight lines and flat walls (polyhedral regions).

**Why the method needs it:** The algorithm's core trick is tweaking the slopes and positions of walls that define an advice region. If an expert gave advice shaped like a sphere or an oval, there are no straight "walls" to tweak, meaning the matrix adjustments ($F_i$) wouldn't have anything to manipulate.

**Violation scenario:** If a doctor says, "At-risk patients have values that fall within a 10-unit radius of this exact average," that creates a circular rule. We'd have to roughly hack it into a box shape first, which could mess up the accuracy before the algorithm even starts refining.

**Paper reference:** Section 2, Equation 2 explicitly defines advice as $D_i x \le d_i$ (which translates to flat polyhedral boundaries).

### Assumption 3

**Assumption:** The expert's advice is mostly correct and only needs a small nudge to be perfect.

**Why the method needs it:** The math formula (Equation 8) has a built-in penalty that forces the $F_i$ and $f_i$ tweaks to stay as small as possible. The model assumes we trust the expert heavily. It assumes we only want to bend the rules a little, not entirely rewrite them.

**Violation scenario:** What if the expert accidentally gave the completely wrong label? Like giving us a rule for "no diabetes" but accidentally labeling it "diabetes". Because the algorithm hates making large adjustments, it would refuse to completely flip the advice around, and the model would get stuck trying to compromise with totally wrong information.

**Paper reference:** Equation 8 includes a $\nu$ penalty to keep refinements small; Algorithm 1 starts by assuming adjustments are 0.

### Assumption 4

**Assumption:** The math solvers will find a "good enough" solution even though the problem isn't perfectly solvable.

**Why the method needs it:** Because the new model adjusts both the data boundary AND the advice boundary at the same time, the math becomes non-convex. This means there isn't one guaranteed perfect answer anymore; there are many "hills and valleys" in the math, and the solver might get stuck in a bad valley. We assume the solver's starting point is good enough to slide into a good valley.

**Violation scenario:** If we have a ton of conflicting expert advice rules that overlap weirdly, the math landscape becomes a total mess. The algorithm might get confused, stop early, and give us a model that performs worse than if we had just used a simple standard SVM.

**Paper reference:** Section 3.1 directly states that Proposition 1 only guarantees reaching a "local minimum" (a valley), not the perfect global minimum.
